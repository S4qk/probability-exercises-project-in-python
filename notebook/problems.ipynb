## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np

Set_X = []
h = 0

for i in range(1000):
    Dice_A = random.randint(1, 6)
    Dice_B = random.randint(1, 6)
    Set_X.append(Dice_A + Dice_B)
    if (Dice_A + Dice_B)%2 == 0 or (Dice_A + Dice_B) > 7:
        h +=1 
    
print(f"The estimated probability that the total score is even or greater than 7 is {(h/1000)*100}%.")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
def ball(A):
    if len(A) == 0:
        return A.append(random.randint(0, 59))
    else:
        for i in range(len(A)):
            while True:
                Nnumber = random.randint(0, 59)
                if Nnumber != A[i]:
                    break
        return A.append(Nnumber)
    
def Tballs(A):
    w, r, g = 0, 0, 0
    for i in range(5):
        if A[i] < 10: #white
            w += 1
        elif A[i] < 30: #red
            r += 1
        else: #green
            g += 1
    if (w == 3) and (r == 2):
        return 2
    elif (w == 5) or (r == 5) or (g == 5):
        return 1
    else:
        return 0

tt, sm = 0, 0

for i in range(1000):
    ChoosenBalls = []
    ball(ChoosenBalls)
    ball(ChoosenBalls)
    ball(ChoosenBalls)
    ball(ChoosenBalls)
    ball(ChoosenBalls)
    if Tballs(ChoosenBalls) == 2:
        tt += 1
    elif Tballs(ChoosenBalls) == 1:
        sm += 1

print(f"The probability of getting 3 white balls and 2 red balls is {np.round((tt/1000)*100, 3)}%.")
print(f"The probability of getting 5 balls with the same color is {np.round((sm/1000)*100, 3)}%.")
```
